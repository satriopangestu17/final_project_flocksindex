```python
import folium
import pandas as pd
import os

states = os.path.join('data', 'customgeo1.geojson.json')
relation_data = os.path.join('data', 'sharperatio.csv')
state_data = pd.read_csv(relation_data)

m = folium.Map(location=[7.7245607,110.5492966], zoom_start=3.6)

m.choropleth(
    geo_data=states,
    name='choropleth',
    data=state_data,
    columns=['countries', 'sharpe ratio'],
    key_on= 'feature.properties.sovereignt',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="sharpe ratio"
)

# Global tooltip
tooltip = 'Click For More Info'
# Create custom marker icon
logoindo = folium.features.CustomIcon('indonesia.png', icon_size=(20, 15))
logolaos = folium.features.CustomIcon('laos1.png', icon_size=(15, 15))
logomalaysia = folium.features.CustomIcon('malaysia1.png', icon_size=(15, 15))
logophilippines = folium.features.CustomIcon('philippines1.png', icon_size=(15, 15))
logosingapore = folium.features.CustomIcon('singapore1.png', icon_size=(15, 15))
logothailand = folium.features.CustomIcon('thailand1.png', icon_size=(15, 15))
logovietnam = folium.features.CustomIcon('vietnam1.png', icon_size=(15, 15))
# Create markers
folium.Marker([19.392157, 102.791406],
              popup='<strong>Laos\nSharpe Ratio: -1.12</strong>',
              tooltip=tooltip,
              icon=logolaos).add_to(m),
folium.Marker([3.127361, 101.685985],
              popup='<strong>Malaysia\nSharpe Ratio: -0.51</strong>',
              tooltip=tooltip,
              icon=logomalaysia).add_to(m),
folium.Marker([14.608167, 120.991732],
              popup='<strong>Philippines\nSharpe Ratio: 0.12</strong>',
              tooltip=tooltip,
              icon=logophilippines).add_to(m),
folium.Marker([1.309866, 103.734333],
              popup='<strong>Singapore\nSharpe Ratio: -0.07</strong>',
              tooltip=tooltip,
              icon=logosingapore).add_to(m),
folium.Marker([14.558128, 100.615295],
              popup='<strong>Thailand\nSharpe Ratio: 0.32</strong>',
              tooltip=tooltip,
              icon=logothailand).add_to(m),
folium.Marker([20.487522, 105.913338],
              popup='<strong>Vietnam\nSharpe Ratio: 0.48</strong>',
              tooltip=tooltip,
              icon=logovietnam).add_to(m),
folium.Marker([-6.228010, 106.854104],
              popup='<strong>Indonesia\nSharpe Ratio: 0.24</strong>',
              tooltip=tooltip,
              icon=logoindo).add_to(m)

folium.LayerControl().add_to(m)
m

# state_data
```
