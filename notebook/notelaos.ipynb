```python
dflaos = dfgab.loc[0:, ['Laos']]
# df.iloc[0:2, 0:2]
# df[['Date', 'Indonesia']][0:2]
dflaos
```

```python
# a variable for prediction 'n' days out into the future
forecast_out = 1
## create another column (the target or dependent variable) shifted 'n' units up
dflaos['predictlaos']= dflaos[['Laos']].shift(-forecast_out)

dflaos
```

```python
## create the independent data set (x)
# convert the dataframe to a numpy array
xlaos = np.array(dflaos.drop(['predictlaos'], axis=1))

xlaos
#remove the 'n' rows
xlaos = xlaos[:-forecast_out]
xlaos
xlaosreshape = xlaos.reshape(-1)
xlaoslist = xlaos.tolist()
# xlaoslist
xlaosreshape
```

```python
## create the dependent data set(y)
# convert dataframe to a numpy array (all of the values includeing the NaN's)

ylaos = np.array(dflaos['predictlaos'])
#get all of the y values except the last 'n' rows
ylaos = ylaos[:-forecast_out]
ylaos
ylaoslist = ylaos.tolist()
ylaoslist
```

```python
dfcorrlaos = pd.DataFrame({'ylaos': ylaoslist, 'xlaos': xlaosreshape})
dfcorrlaos
dfcorrlaos = dfcorrlaos.corr()
dfcorrlaos
```

```python
from sklearn.model_selection import train_test_split
# xtr, xts, ytr, yts = train_test_split(df[['JCI']][:-forecast_out], df['prediction'][:-forecast_out], random_state=10)
xtrlaos, xtslaos, ytrlaos, ytslaos = train_test_split(xlaos, ylaos, random_state=5)
xtrlaos
```

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    rmse = np.sqrt(mean_squared_error(test_labels, predictions))
    r2 = r2_score(test_labels, predictions)
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.4f}%.'.format(accuracy))
    print('RMSE = {:0.4f}'.format(rmse))
    print('R2 Score = {:0.4f}.'.format(r2))
    vals = [accuracy,rmse,r2]
    return vals
```

```python
lnr_grid = {
    'fit_intercept': [True, False],
    'normalize': [True, False],
    'copy_X': [True, False],
    'n_jobs':['int', None]
}
print(lnr_grid)

```

```python
lnr = LinearRegression()
lnr_random = RandomizedSearchCV(estimator=lnr, param_distributions=lnr_grid, n_iter=100,
                               cv=5, verbose=2, random_state=41, n_jobs=-1)

lnr_random.fit(xtrlaos, ytrlaos)
```

```python
lnr_random.best_params_
```

```python
## LinearRegression with default hyperparameters

base_lnr = LinearRegression()
base_lnr.fit(xtrlaos, ytrlaos)
base_lnr_accuracy = evaluate(base_lnr, xtslaos, ytslaos)
base_lnr_accuracy
```

```python
## LinearRegression with default best hyperparameters

best_lnr = lnr_random.best_estimator_
best_lnr.fit(xtrlaos, ytrlaos)
best_lnr_accuracy = evaluate(best_lnr, xtrlaos, ytrlaos)
```
