```python
## convert to US dollar
dfkali = df.loc[0:, 'Indonesia':]
df2kali = df2.loc[0:, 'Indonesia':]
dfgab = dfkali / df2kali
dfgab
```

```python
dfphilip = dfgab.loc[0:, ['Philippines']]
# df.iloc[0:2, 0:2]
# df[['Date', 'Indonesia']][0:2]
dfphilip
```

```python
# a variable for prediction 'n' days out into the future
forecast_out = 1
## create another column (the target or dependent variable) shifted 'n' units up
dfphilip['predictphilippines']= dfphilip[['Philippines']].shift(-forecast_out)

dfphilip
```

```python
## create the independent data set (x)
# convert the dataframe to a numpy array
xphilip = np.array(dfphilip.drop(['predictphilippines'], axis=1))

xphilip
#remove the 'n' rows
xphilip = xphilip[:-forecast_out]
xphilip
xphilipreshape = xphilip.reshape(-1)
xphilipreshape
```

```python
## create the dependent data set(y)
# convert dataframe to a numpy array (all of the values includeing the NaN's)

yphilip = np.array(dfphilip['predictphilippines'])
#get all of the y values except the last 'n' rows
yphilip = yphilip[:-forecast_out]
yphilip
```

```python
dfcorrphilip = pd.DataFrame({'yphilip': yphilip, 'xphilip': xphilipreshape})
dfcorrphilip
dfcorrphilip = dfcorrphilip.corr()
dfcorrphilip

```

```python
from sklearn.model_selection import train_test_split
# xtr, xts, ytr, yts = train_test_split(df[['JCI']][:-forecast_out], df['prediction'][:-forecast_out], random_state=10)
xtrphilip, xtsphilip, ytrphilip, ytsphilip = train_test_split(xphilip, yphilip, random_state=5)
xtrphilip
```

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    rmse = np.sqrt(mean_squared_error(test_labels, predictions))
    r2 = r2_score(test_labels, predictions)
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.4f}%.'.format(accuracy))
    print('RMSE = {:0.4f}'.format(rmse))
    print('R2 Score = {:0.4f}.'.format(r2))
    vals = [accuracy, rmse, r2]
    return vals
```

```python
lnr_grid = {
    'fit_intercept': [True, False],
    'normalize': [True, False],
    'copy_X': [True, False],
    'n_jobs':['int', None]
}
print(lnr_grid)
```

```python
lnr = LinearRegression()
lnr_random = RandomizedSearchCV(estimator=lnr, param_distributions=lnr_grid, n_iter=100,
                               cv=5, verbose=2, random_state=41, n_jobs=-1)

lnr_random.fit(xtrphilip, ytrphilip)
```

```python
lnr_random.best_params_
```

```python
## LinearRegression with default hyperparameters

base_lnr = LinearRegression()
base_lnr.fit(xtrphilip, ytrphilip)
base_lnr_accuracy = evaluate(base_lnr, xtsphilip, ytsphilip)
```

```python
## LinearRegression with default best hyperparameters

best_lnr = lnr_random.best_estimator_
best_lnr.fit(xtrphilip, ytrphilip)
best_lnr_accuracy = evaluate(best_lnr, xtsphilip, ytsphilip)
best_lnr_accuracy
```
